## 读取某一支股票的book数据
该数据的存储格式为parquet，用pandas下的read_parquet函数进行去读取。
可能会报错：
```
Unable to find a usable engine; tried using: 'pyarrow', 'fastparquet'.
```
提示read_parquet函数缺少对应的引擎，通过conda install缺少的包修复这一问题。
```
conda install -c conda-forge pyarrow
conda install -c conda-forge fastparquet
```

````python
# 指定要读取的文件的路径
root_dir = '../data'
dataDir = os.path.join(root_dir, 'book_train.parquet/stock_id=0/c439ef22282f412ba39e9137a3fdabac.parquet')
# 通过pandas读取文件
df_stockBook =  pd.read_parquet(dataDir)
# 展示该表格数据的前10行
df_stockBook.head(10)
````
